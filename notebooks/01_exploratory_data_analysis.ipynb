# Student Performance Prediction - Exploratory Data Analysis
## Astana IT University - Educational Practice

This notebook contains exploratory data analysis for the global university students performance and habits dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Load Dataset

```python
df = pd.read_csv('../data/global_university_students_performance_habits_10000.csv')
print(f"Dataset shape: {df.shape}")
df.head()
```

## 2. Dataset Overview

```python
df.info()
```

```python
df.describe()
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

## 3. Target Variables Analysis
Our main targets are `final_exam_score` and `assignment_score`

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(df['final_exam_score'], bins=30, edgecolor='black', alpha=0.7)
axes[0].set_title('Final Exam Score Distribution')
axes[0].set_xlabel('Score')
axes[0].set_ylabel('Frequency')

axes[1].hist(df['assignment_score'], bins=30, edgecolor='black', alpha=0.7, color='orange')
axes[1].set_title('Assignment Score Distribution')
axes[1].set_xlabel('Score')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
print("Target Variables Statistics:")
print(df[['final_exam_score', 'assignment_score', 'GPA']].describe())
```

## 4. Feature Analysis

```python
categorical_cols = ['gender', 'country', 'major', 'part_time_job', 'relationship_status', 
                   'family_income_level', 'internet_quality', 'favorite_AI_tool', 'note_taking_method']

print("Categorical Features:")
for col in categorical_cols:
    if col in df.columns:
        print(f"\n{col}: {df[col].nunique()} unique values")
        print(df[col].value_counts().head())
```

## 5. Study Habits vs Performance

```python
study_features = ['study_hours_per_day', 'class_attendance_percent', 'exam_preparation_days']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, feature in enumerate(study_features):
    axes[i].scatter(df[feature], df['final_exam_score'], alpha=0.5)
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Final Exam Score')
    axes[i].set_title(f'{feature} vs Final Exam Score')

plt.tight_layout()
plt.show()
```

## 6. Lifestyle Factors vs Performance

```python
lifestyle_features = ['sleep_hours', 'screen_time_hours', 'social_media_hours', 
                     'gaming_hours', 'exercise_hours_per_week', 'coffee_consumption_per_day']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.ravel()

for i, feature in enumerate(lifestyle_features):
    axes[i].scatter(df[feature], df['final_exam_score'], alpha=0.5, color='green')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Final Exam Score')
    axes[i].set_title(f'{feature} vs Performance')

plt.tight_layout()
plt.show()
```

## 7. AI Tool Usage Analysis

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

df.groupby('favorite_AI_tool')['final_exam_score'].mean().sort_values().plot(kind='barh', ax=axes[0])
axes[0].set_title('Average Final Exam Score by AI Tool')
axes[0].set_xlabel('Average Score')

axes[1].scatter(df['AI_tool_usage_hours'], df['final_exam_score'], alpha=0.5, color='purple')
axes[1].set_xlabel('AI Tool Usage Hours')
axes[1].set_ylabel('Final Exam Score')
axes[1].set_title('AI Usage vs Performance')

plt.tight_layout()
plt.show()
```

## 8. Correlation Analysis

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols = [col for col in numeric_cols if col != 'student_id']

correlation_matrix = df[numeric_cols].corr()

plt.figure(figsize=(16, 14))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Feature Correlation Matrix', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```

```python
print("Top correlations with final_exam_score:")
print(correlation_matrix['final_exam_score'].sort_values(ascending=False).head(10))

print("\nTop correlations with assignment_score:")
print(correlation_matrix['assignment_score'].sort_values(ascending=False).head(10))
```

## 9. Mental Stress Analysis

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].scatter(df['mental_stress_level'], df['final_exam_score'], alpha=0.5, color='red')
axes[0].set_xlabel('Mental Stress Level')
axes[0].set_ylabel('Final Exam Score')
axes[0].set_title('Mental Stress vs Performance')

stress_by_major = df.groupby('major')['mental_stress_level'].mean().sort_values(ascending=False).head(10)
stress_by_major.plot(kind='barh', ax=axes[1], color='coral')
axes[1].set_title('Average Mental Stress by Major (Top 10)')
axes[1].set_xlabel('Average Stress Level')

plt.tight_layout()
plt.show()
```

## 10. GPA vs Final Performance

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].scatter(df['GPA'], df['final_exam_score'], alpha=0.5)
axes[0].set_xlabel('GPA')
axes[0].set_ylabel('Final Exam Score')
axes[0].set_title('GPA vs Final Exam Score')

axes[1].scatter(df['GPA'], df['assignment_score'], alpha=0.5, color='orange')
axes[1].set_xlabel('GPA')
axes[1].set_ylabel('Assignment Score')
axes[1].set_title('GPA vs Assignment Score')

plt.tight_layout()
plt.show()
```

## 11. Key Insights Summary

```python
print("="*60)
print("KEY INSIGHTS FROM EXPLORATORY DATA ANALYSIS")
print("="*60)

print(f"\n1. Dataset Size: {len(df)} students from {df['country'].nunique()} countries")
print(f"2. Number of majors: {df['major'].nunique()}")
print(f"3. Average Final Exam Score: {df['final_exam_score'].mean():.2f}")
print(f"4. Average Assignment Score: {df['assignment_score'].mean():.2f}")
print(f"5. Average GPA: {df['GPA'].mean():.2f}")
print(f"6. Students using AI tools: {df['favorite_AI_tool'].notna().sum()} ({df['favorite_AI_tool'].notna().sum()/len(df)*100:.1f}%)")
print(f"7. Students with part-time jobs: {(df['part_time_job'] == 'Yes').sum()} ({(df['part_time_job'] == 'Yes').sum()/len(df)*100:.1f}%)")

print("\nTop 5 Most Common Majors:")
print(df['major'].value_counts().head())

print("\nMost Popular AI Tools:")
print(df['favorite_AI_tool'].value_counts())
```
